### [警察庁オープンデータ](https://www.npa.go.jp/publications/statistics/koutsuu/opendata/index_opendata.html)で公表されている2019年〜2023年の各本票のcsvファイルを結合して、保存する

各データは以下の通り：   
[２０１９年](https://www.npa.go.jp/publications/statistics/koutsuu/opendata/2019/opendata_2019.html)   
[２０２０年](https://www.npa.go.jp/publications/statistics/koutsuu/opendata/2020/opendata_2020.html)   
[２０２１年](https://www.npa.go.jp/publications/statistics/koutsuu/opendata/2021/opendata_2021.html)   
[２０２２年](https://www.npa.go.jp/publications/statistics/koutsuu/opendata/2022/opendata_2022.html)   
[２０２３年](https://www.npa.go.jp/publications/statistics/koutsuu/opendata/2023/opendata_2023.html)

```python
import pandas as pd
```

```python
df_2019 = pd.read_csv("honhyo_2019.csv", encoding="shift-jis")
```

```python
df_2020 = pd.read_csv("honhyo_2020.csv", encoding="shift-jis")
```

```python
df_2021 = pd.read_csv("honhyo_2021.csv", encoding="shift-jis")
```

```python
df_2022 = pd.read_csv("honhyo_2022.csv", encoding="shift-jis")
```

```python
df_2023 = pd.read_csv("honhyo_2023.csv", encoding="shift-jis")
```

```python
print(df_2019.columns)
```

```python
print(df_2023.columns)
```

```python
# 2023にあって、2019にないカラムを表示する
print(list(filter(lambda x: x not in df_2019.columns,  df_2023.columns)))
```

```python
# 2019にあって、2023にないカラムを表示する
print(list(filter(lambda x: x not in df_2023.columns,  df_2019.columns)))
```

```python
# 各データフレームのカラム名をセットに変換
columns_df_2019 = set(df_2019.columns)
columns_df_2020 = set(df_2020.columns)
columns_df_2021 = set(df_2021.columns)
columns_df_2022 = set(df_2022.columns)
columns_df_2023 = set(df_2023.columns)
```

```python
# 全てのデータフレームに共通するカラムを見つける
common_columns = columns_df_2019 & columns_df_2020 & columns_df_2021 & columns_df_2022 & columns_df_2023

print(common_columns)
```

```python
# 全てのカラムの和集合を求める
all_columns = columns_df_2019 | columns_df_2020 | columns_df_2021 | columns_df_2022 | columns_df_2023

# 共通しないカラムを特定する
non_common_columns = all_columns - common_columns

print("共通しないカラム:", non_common_columns)
```

```python
non_common_columns
```

```python
# データフレームのリストを作成
dfs = [df_2019, df_2020, df_2021, df_2022, df_2023]
```

```python
# 共通カラムのリストに変換
common_columns = list(common_columns)

print("共通するカラム:", common_columns)
```

```python
# 全てのデータフレームを共通のカラムで外部結合

from functools import reduce

result = reduce(lambda left, right: pd.merge(left, right, on=common_columns, how='outer'), dfs)
```

```python
result
```

```python
# エントリー数を確認する
len(df_2019)+len(df_2020)+len(df_2021)+len(df_2022)+len(df_2023)
```

```python
# 外部結合した全データフレームをCSVファイルとして保存
result.to_csv('2019-2023df.csv', index=False)
```

### 全データから電動アシスト自転車のデータを抽出して保存する

```python
# 電動アシスト自転車の事故だけ抽出する
result.columns
```

```python
electric_bike_df = result[(result['当事者種別（当事者A）'] == 52) | (result['当事者種別（当事者B）'] == 52)]
```

```python
electric_bike_df
```

```python
# 電動アシスト自転車のみのデータフレームをCSVファイルとして保存
electric_bike_df.to_csv('電動アシスト自転車df.csv', index=False)
```
